 ## **Column Descriptions:**


The dataset contains 112,457 rows and 8 columns. Below is an explanation of each column:


Index:

- Represents the stock market index or the specific stock symbol.

Date:

- Indicates the date for each record, which may need conversion to a proper datetime format.

Open:

- The price at which the stock or market opened on the given date.

High:

- The highest price reached by the stock or market on that date.

Low:

- The lowest price recorded for the stock or market on that date.

Close:

- The price at which the stock or market closed on that date.

Adj Close:

- The adjusted closing price, which accounts for dividends and stock splits.

Volume:

- The total number of shares traded on that particular day.
